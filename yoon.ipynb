```python
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

import os
import pandas as pd
import re

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', -1)
```

```python
chrome_options = Options()
# chrome_options.add_argument('--headless')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

url = 'https://foster.uw.edu/faculty-research/academic-departments/accounting/faculty/'
driver.get(url)
```

```python
df = pd.DataFrame(columns=[['Name', 'Status', 'Emerit', 'Assistant', 'Associate', 'Lecturer&Adjunct&Fellow&Visiting']])

soup = BeautifulSoup(driver.page_source, 'html.parser')
temp = soup.find_all('div', class_='section group', id='list')

for i, j in enumerate(temp[0].find_all('div', class_='col span_6_of_12')):
    name = j.find('h2', class_= 'entry-title').text
    status = j.find('h3', class_= 'entry-title').text
    
    df.loc[i, 'Name'] = name
    df.loc[i, 'Status'] = status
    
for i in df.index:
    status = str(df.loc[i, 'Status'])
    
    if "Emerit" in status:
        df.loc[i, 'Emerit'] = 'Yes'
    if 'Assistant' in status:
        df.loc[i, 'Assistant'] = 'Yes'
    if 'Associate' in status:
        df.loc[i, 'Associate'] = 'Yes'
        
    if 'Lecturer' in status:
        df.loc[i, 'Lecturer&Adjunct&Fellow&Visiting'] = 'Yes'
        
    if 'Adjunct' in status:
        df.loc[i, 'Lecturer&Adjunct&Fellow&Visiting'] = 'Yes'
        
    if 'Fellow' in status:
        df.loc[i, 'Lecturer&Adjunct&Fellow&Visiting'] = 'Yes'
        
    if 'Visiting' in status:
        df.loc[i, 'Lecturer&Adjunct&Fellow&Visiting'] = 'Yes'
        
df = df.fillna('0')
df
```

```python
df = pd.DataFrame(columns=[['Name', 'Status', 'Emerit', 'Assistant', 'Associate', 'Lecturer&Adjunct&Fellow&Visiting', 'url']])

soup = BeautifulSoup(driver.page_source, 'html.parser')
temp = soup.find_all('div', class_='col span_12_of_12 slide pop')

index = 0
for i in temp:
    i = i.find_all('a')
    for a, num in enumerate(i):
        name = num.text
        url = str(num).split('href="')[1]
        url = url.split('">')[0]
        
        df.loc[index, 'Name'] = name
        df.loc[index, 'url'] = url
        index = index + 1
        
for i in df.index:
    url = df.iloc[i, 6]
    driver.get(url)
    time.sleep(1)
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    temp = soup.find('ul', class_='faculty-profile__list')
    temp = temp.text.split('\n')[1]
    df.loc[i, 'Status'] = temp
    
for i in df.index:
    status = str(df.loc[i, 'Status'])
    
    if "Emerit" in status:
        df.loc[i, 'Emerit'] = 'Yes'
    if 'Assistant' in status:
        df.loc[i, 'Assistant'] = 'Yes'
    if 'Associate' in status:
        df.loc[i, 'Associate'] = 'Yes'
        
    if 'Lecturer' in status:
        df.loc[i, 'Lecturer&Adjunct&Fellow&Visiting'] = 'Yes'
        
    if 'Adjunct' in status:
        df.loc[i, 'Lecturer&Adjunct&Fellow&Visiting'] = 'Yes'
        
    if 'Fellow' in status:
        df.loc[i, 'Lecturer&Adjunct&Fellow&Visiting'] = 'Yes'
        
    if 'Visiting' in status:
        df.loc[i, 'Lecturer&Adjunct&Fellow&Visiting'] = 'Yes'
        
df = df.fillna('0')
df = df.drop('url', axis=1)
df
```
